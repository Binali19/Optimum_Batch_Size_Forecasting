# Importing the libraries

```python
#importing needed packages
%matplotlib inline
from matplotlib import pyplot as plt
#import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.mlab as mlab
```

# Converting into a numpy array

```python

df=df.as_matrix() 
```

# The dataset is divided in data and target

```python
data=df[:,0:3]
target=df[:,3]

print(target)
```

# example data

```python

mu = 100 # mean of distribution

num_bins = 4
x=target
# the histogram of the data
n, bins, patches = plt.hist(x, num_bins, normed=1, facecolor='blue', alpha=0.5)

plt.xlabel('Optimum batch size (% Range)')
plt.ylabel('Probability')

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
plt.show()
```

# Splitting Original Data set into Train set and Test set

```python
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import KFold
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
import numpy as np

target=np.reshape(target, (-1,1))
scaler_x = MinMaxScaler()
scaler_y = MinMaxScaler()
scaler_x.fit(data)
scaler_y.fit(target)
xscale=scaler_x.transform(data)
yscale=scaler_y.transform(target)

train_data, test_data, train_target, test_target = train_test_split(xscale, yscale,test_size=0.3)
```

# Feedforward Neural Network Model

```python
import keras.models as models
import keras.layers as layers
import keras.optimizers as optimizers
from keras.layers import Dropout
import numpy as np

model = models.Sequential()
model.add(layers.Dense(128, input_dim=3, kernel_initializer='normal', activation='relu'))
model.add(Dropout(0.2))
model.add(layers.Dense(128, activation='relu'))
model.add(Dropout(0.2))
model.add(layers.Dense(10, activation='relu'))
model.add(layers.Dense(1, activation='linear'))

model.compile(optimizer='adam',loss='mse',metrics=['mse','mae'])

model.summary()
```

# Training the dataset

```python
history=model.fit(train_data,train_target,epochs=100,validation_split=0.2,batch_size=32)
```

# Plotting the model value loss 

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()
```

# R2 Score

```python
from sklearn.metrics import r2_score

results=model.predict(test_data)

r2=r2_score(test_target,results)
print('r2 score:',r2)
r2=r2_score(test_target,results)*100
print('r2 score:',r2)
```

# Saving the trained weights

```python
model.save('Optmium_batch_size.h5')
```
